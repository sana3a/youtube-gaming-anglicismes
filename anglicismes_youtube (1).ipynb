```python
from googleapiclient.discovery import build
from bs4 import BeautifulSoup
import re

#paramètres
api_key = "CLE API" # Clé API personnelle non fournie pour des raisons de sécurité
video_ids = ["CM-kN9xs3Uw", "89AeQnSzfl0"]  #les ID des vidéos
max_comments = 100

#Initialisation API
youtube = build('youtube', 'v3', developerKey=api_key)

all_comments = []


for video_id in video_ids:
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=max_comments,
        textFormat="plainText"
    )
    response = request.execute()

    
    for item in response['items']:
        com = item['snippet']['topLevelComment']['snippet']['textDisplay']
        all_comments.append(com)

print(f"Total commentaires récupérés : {len(all_comments)}")

#Nettoyage/Tokenisation 
tokens = []
for com in all_comments:
    com = BeautifulSoup(com, "html.parser").get_text()   #enlève HTML
    mots = re.findall(r'\b\w+\b', com)                   #tokenisation simple
    tokens.extend(mots)

print("\nExemple de tokens :")
print(tokens[:1000])
```

```python
import fasttext
import re

#Charger le modèle FastText
model = fasttext.load_model("lid.176.ftz")

#Nettoyage des tokens 
clean_tokens = []
for token in tokens:

    #Retirer les balises HTML
    if token.startswith("<"):
        continue

    #Retirer les nombres
    if token.isdigit():
        continue

    #Nettoyage 
    token_clean = re.sub(r"[^\w\-']+", "", token)

    #Retirer les tokens trop courts
    if len(token_clean) <= 1:
        continue

    clean_tokens.append(token_clean)

#Supression des doublons
clean_tokens = list(set(clean_tokens))  

#Détection de langue
languages = []
for token in clean_tokens:
    prediction = model.predict(token.lower())
    lang_code = prediction[0][0].replace("__label__", "")
    languages.append((token, lang_code))

#Filtrer les tokens anglais
english_tokens = [t for t, lang in languages if lang == "en"]
french_tokens = [t for t, lang in languages if lang == "fr"]


print("Tokens probablement anglais :\n", english_tokens)
print("Tokens probablement français :\n", french_tokens)

print("\nToutes les détections :\n", languages)
```
